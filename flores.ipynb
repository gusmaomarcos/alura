```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])

eixo.scatter(x=df['comprimento_sépala'], y=df['largura_sépala'])
eixo.set_title('Gráfico de Dispersão')
eixo.set_xlabel('Comprimento da Sépala')
eixo.set_ylabel('Largura da Sépala')
```

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])

eixo.scatter(x=df['comprimento_sépala'], y=df['largura_sépala'])
eixo.set_title('Gráfico de Dispersão',fontsize=16)
eixo.set_xlabel('Comprimento da Sépala',fontsize=14)
eixo.set_ylabel('Largura da Sépala', fontsize=14)
eixo.tick_params(labelsize=12)
```

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])

cores = {'Iris-setosa':'r', 'Iris-versicolor':'b', 'Iris-virginica':'g'}

for especie in df['espécie'].unique():
    tmp = df[df['espécie'] == especie]
    eixo.scatter(tmp['comprimento_sépala'], tmp['largura_sépala'],
                color=cores[especie])

#eixo.scatter(x=df['comprimento_sépala'], y=df['largura_sépala'])
eixo.set_title('Gráfico de Dispersão',fontsize=16)
eixo.set_xlabel('Comprimento da Sépala',fontsize=14)
eixo.set_ylabel('Largura da Sépala', fontsize=14)
eixo.tick_params(labelsize=12)
eixo.legend(cores, fontsize=12)
```

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])

cores = {'Iris-setosa':'r', 'Iris-versicolor':'b', 'Iris-virginica':'g'}
marcadores = {'Iris-setosa':'x', 'Iris-versicolor':'o', 'Iris-virginica':'v'}

for especie in df['espécie'].unique():
    tmp = df[df['espécie'] == especie]
    eixo.scatter(tmp['comprimento_sépala'], tmp['largura_sépala'],
                color=cores[especie], marker=marcadores[especie])

eixo.set_title('Gráfico de Dispersão',fontsize=16)
eixo.set_xlabel('Comprimento da Sépala',fontsize=14)
eixo.set_ylabel('Largura da Sépala', fontsize=14)
eixo.tick_params(labelsize=12)
eixo.legend(cores, fontsize=12)
```

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])

cores = {'Iris-setosa':'r', 'Iris-versicolor':'b', 'Iris-virginica':'g'}
marcadores = {'Iris-setosa':'x', 'Iris-versicolor':'o', 'Iris-virginica':'v'}

for especie in df['espécie'].unique():
    tmp = df[df['espécie'] == especie]
    eixo.scatter(tmp['comprimento_sépala'], tmp['largura_sépala'],
                color=cores[especie], marker=marcadores[especie],
                s=100)

eixo.set_title('Gráfico de Dispersão',fontsize=16)
eixo.set_xlabel('Comprimento da Sépala',fontsize=14)
eixo.set_ylabel('Largura da Sépala', fontsize=14)
eixo.tick_params(labelsize=12)
eixo.legend(cores, fontsize=12)
```

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

mu, sigma = df['comprimento_pétala'].mean(), df['comprimento_pétala'].std()

eixo.hist(df['comprimento_pétala'], bins=20)

eixo.set_title('Gráfico de Caixa',fontsize=16)
eixo.set_xlabel('Comprimento da Pétala')
eixo.grid(True)

eixo.annotate('$mu = {0:.2f}$\n$sigma = {1:.2f}$'.format(mu, sigma),
             xy=(5.1, 28), fontsize=16)
eixo.axvline(mu, color='k', linestyle='--')
eixo.annotate('Média', xy=(mu-1, 20.2), fontsize=16)
```

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

mu, sigma = df['comprimento_pétala'].mean(), df['comprimento_pétala'].std()

eixo.hist(df['comprimento_pétala'], bins=20)

eixo.set_title('Gráfico de Caixa',fontsize=16)
eixo.set_xlabel('Comprimento da Pétala')
eixo.grid(True)

eixo.annotate('$\mu = {0:.2f}$\n$\sigma = {1:.2f}$'.format(mu, sigma),
             xy=(5.1, 28), fontsize=16)
eixo.axvline(mu, color='k', linestyle='--')
eixo.annotate('Média', xy=(mu-1, 20.2), fontsize=16)

eixo.axvline(df['comprimento_pétala'].median(), color='r', linestyle='--')
eixo.annotate('Mediana', xy=(mu+.7, 20.2), fontsize=16)
```

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

df_iv = df[df['espécie'] == 'Iris-versicolor']

mu, sigma = df_iv['comprimento_pétala'].mean(), df_iv['comprimento_pétala'].std()

eixo.hist(df_iv['comprimento_pétala'], bins=20)

eixo.set_title('Iris Versicolor',fontsize=16)
eixo.set_xlabel('Comprimento da Pétala')
eixo.grid(True)

eixo.annotate('$\mu = {0:.2f}$\n$\sigma = {1:.2f}$'.format(mu, sigma),
             xy=(4.6, 6), fontsize=16)
eixo.axvline(mu, color='k', linestyle='--')
eixo.annotate('Média', xy=(mu-.4, 5.5), fontsize=16)

eixo.axvline(df_iv['comprimento_pétala'].median(), color='r', linestyle='--')
eixo.annotate('Mediana', xy=(mu-.5, 6.5), fontsize=16, color='r')

fig.savefig('Histograma_iv.png', bbox_inches='tight')
```

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

df_is = df[df['espécie'] == 'Iris-setosa']

mu, sigma = df_is['comprimento_pétala'].mean(), df_is['comprimento_pétala'].std()

eixo.hist(df_is['comprimento_pétala'], bins=20)

eixo.set_title('Iris Setosa',fontsize=16)
eixo.set_xlabel('Comprimento da Pétala')
eixo.grid(True)

eixo.annotate('$\mu = {0:.2f}$\n$\sigma = {1:.2f}$'.format(mu, sigma),
             xy=(1.7, 7), fontsize=16)
eixo.axvline(mu, color='k', linestyle='--')
eixo.annotate('Média', xy=(mu-.15, 13.5), fontsize=16)

eixo.axvline(df_is['comprimento_pétala'].median(), color='r', linestyle='--')
eixo.annotate('Mediana', xy=(mu+.08, 12), fontsize=16, color='r')

fig.savefig('Histograma_is.png', bbox_inches='tight')
```

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

df_ig = df[df['espécie'] == 'Iris-virginica']

mu, sigma = df_ig['comprimento_pétala'].mean(), df_ig['comprimento_pétala'].std()

eixo.hist(df_ig['comprimento_pétala'], bins=20)

eixo.set_title('Iris Virginica',fontsize=16)
eixo.set_xlabel('Comprimento da Pétala')
eixo.grid(True)

eixo.annotate('$\mu = {0:.2f}$\n$\sigma = {1:.2f}$'.format(mu, sigma),
             xy=(6.3, 7), fontsize=16)
eixo.axvline(mu, color='k', linestyle='--')
eixo.annotate('Média', xy=(mu, 8), fontsize=16)

eixo.axvline(df_ig['comprimento_pétala'].median(), color='r', linestyle='--')
eixo.annotate('Mediana', xy=(mu, 7), fontsize=16, color='r')

fig.savefig('Histograma_ig.png', bbox_inches='tight')
```
